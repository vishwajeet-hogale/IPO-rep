```
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```


url = 'https://www.fr.de/suche/?tt=1&tx=&sb=&td=&fd=&qr=IPO'

url1 = 'https://www.fr.de'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

#print(soup)



links = []
print(soup.find_all('a', {'class': 'id-LinkOverlay-link'}))
for i in soup.find_all('a', {'class': 'id-LinkOverlay-link'}):
  # for i in a.find_all('a',href=True):
    #print(i['href'])
    links.append(url1 + i["href"])
    print(i)

#links.append(a["href"])
#print(links)

title = []
date = []
s_date = []
text = []

for l in links:
  try:
    fetch = requests.get(l)
    
    sp = BeautifulSoup(fetch.content, 'html.parser')
    
    x=sp.find("h2", { "class" : "id-Article-headline" }).text
    #print(x)

      
      
    y = sp.find("span", { "class" : "id-DateTime-date" }).text
    #print(y)

    
    z = sp.find("div", { "class" : "id-Article-body lp_article_content" }).text
  #print(z)
  
    date.append(y)
    title.append(x)
    text.append(z)
  except:
    continue
  


df2 = pd.DataFrame(list(zip(title,links,date,text)), 
        columns =['title', 'link','publish_date','text'])


df2.to_csv('german1.csv')
```
