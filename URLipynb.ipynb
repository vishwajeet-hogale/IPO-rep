```python
import pandas as pd
import requests 
```

```python
df = pd.read_csv("FinalReport_2021-06-17.csv")
```

```python
invalid_links = []
for i in df["link"].tolist():
  response = requests.get(df["link"][0]);
  if(response.status_code != 200):
    invalid_links.append(str(i))
  print(response)
```
