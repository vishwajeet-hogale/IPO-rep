```python
from datetime import datetime
curr_time=datetime.now()
curr_time=str(curr_time)
curr_time=curr_time.split(' ')
curr_time=curr_time[0]
curr_time=curr_time.split('-')
orig_day=curr_time[2]
orig_mon=curr_time[1]
orig_year=curr_time[0]
print(orig_day)
print(orig_mon)
print(orig_year)
if (orig_mon=='01'):
  orig_mon='Jan'
elif (orig_mon=='02'):
  orig_mon='Feb' 
elif (orig_mon=='03'):
  orig_mon='Mar' 
elif (orig_mon=='04'):
  orig_mon='Apr' 
elif (orig_mon=='05'):
  orig_mon='May' 
elif (orig_mon=='06'):
  orig_mon='Jun' 
elif (orig_mon=='07'):
  orig_mon='Jul' 
elif (orig_mon=='08'):
  orig_mon='Aug' 
elif (orig_mon=='09'):
  orig_mon='Sep' 
elif (orig_mon=='10'):
  orig_mon='Oct' 
elif (orig_mon=='11'):
  orig_mon='Nov' 
elif (orig_mon=='12'):
  orig_mon='Dec' 
sys_date=orig_day+ " "+orig_mon+" "+orig_year
print(sys_date) 

```

```python
import html

def get_frame(title,link,published_date,scraped_date):
  df1 = pd.DataFrame(list(zip(title,link,published_date,scraped_date)), 
               columns =['title','link','published_date','scraped_date'])
  return df1

def convert_utf(title):
   

  for i in range(0,len(title)):
     title[i]=html.unescape(title[i])
  return title
def get_home(link):
   for i in range(0,len(link)):
      #print(df1['link'][i])
      response=requests.get(link[i])
      link[i]=response.url
   return link



if len(title)==0:
  title.append(soup.find_all('title')[1].text)
  link.append(soup.find_all('guid')[0].text)
  scraped_date.append(sys_date)
  published_date.append(soup.find_all('pubdate')[0].text)
  title=convert_utf(title)
  #link=get_home(link)
  fin_df=get_frame(title,link,published_date,scraped_date)
else:
  title=convert_utf(title)
  #link=get_home(link)
  fin_df=get_frame(title,link,published_date,scraped_date)
  
```
