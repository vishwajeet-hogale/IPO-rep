```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python


url = 'https://www.tert.am/am/news/search?q=IPO'

url1 = 'https://www.tert.am'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

# print(soup)



links = []
print(soup.find_all('a', {'class': 'list__link db'}))
for i in soup.find_all('a', {'class': 'list__link db'}):
    if(i["href"].startswith("/")):
      links.append(url1+i["href"])
    else:
      links.append(i["href"])
    print(i["href"])

# links.append(a["href"])
# print(links)

title = []
date = []
s_date = []
text = []

for l in links:
  try:
    # print(l)
    fetch = requests.get(l)
    
    sp = BeautifulSoup(fetch.content, 'html.parser')
    # print(sp)
    x=sp.find("h1", { "class" : "inner-content__article-title fb fs20" }).text
    # print(x)

      
      
    y = sp.find("div", { "class" : "inner-content__article-date fb fs12" }).text
    # print(y)

    
    z = sp.find("div", { "id" : "news-content-container" }).text
    # print(z)
  
    date.append(y)
    title.append(x)
    text.append(z)
  except:
    continue
  


df2 = pd.DataFrame(list(zip(title,links,date,text)), 
        columns =['title', 'link','publish_date','text'])


df2.to_csv('armenia.csv')
```
