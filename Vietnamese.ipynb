```
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```


url = 'https://www.tienphong.vn/search/SVBP/ipo.tpo?st=1&z=0'

url1 = 'https://www.tienphong.vn'
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

# print(soup)



links = []
# print(soup.find_all('article', {'class': 'art_item'}))
for i in soup.find_all('a', {'class': 'story-link'}):
    if(i["href"].startswith("/")):
      links.append(url1+i["href"])
    else:
      links.append(i["href"])
    # print(i)

# links.append(a["href"])
# print(links)

title = []
date = []
s_date = []
text = []

for l in links:
  try:
    # print(l)
    fetch = requests.get(l)
    
    sp = BeautifulSoup(fetch.content, 'html.parser')
    # print(sp)
    x=sp.find("h1", { "class" : "headline cms-title" }).text
    # print(x)

      
      
    y = sp.find("p", { "class" : "byline-dateline" }).time.text
    # print(y)

    
    z = sp.find("div", { "id" : "article-body" }).text
    # print(z)
  
    date.append(y)
    title.append(x)
    text.append(z)
  except:
    continue
  


df2 = pd.DataFrame(list(zip(title,links,date,text)), 
        columns =['title', 'link','publish_date','text'])


df2.to_csv('vietnamese.csv')
```
