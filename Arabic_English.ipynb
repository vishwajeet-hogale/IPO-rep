```python
def convert_to_arabic(df):
  translator = Translator()
  new_list = []
  for i in df["text"]:
    translations = translator.translate(i,dest="en")
    print(translations.text)
    new_list.append(str(translations.text))
  df["converted_text"] = new_list
  return df
    

```

```python
import pandas as pd
df = pd.read_csv("romania.csv")
df = convert_to_arabic(df)
```
